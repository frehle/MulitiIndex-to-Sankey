```python
import pandas as pd
import numpy as np
```

```python
#import raw Statistics Canada physical flow account for plastic material, by product category, 
#from .csv, release date: 2024-03-18
data_raw_StatCan_prod = pd.read_csv('../Raw_Data/38100150.csv',sep=',')
```

```python
data_raw_StatCan_prod
```

```python
#create DataFrame from raw data
df_StatCan_prod=pd.DataFrame(data=data_raw_StatCan_prod)
df_StatCan_prod
```

```python
#remove unnecessary columns
df_StatCan_prod=pd.DataFrame(df_StatCan_prod, columns=['REF_DATE','GEO','Variable','Product category',
                                                       'UOM','COORDINATE','VALUE'])
df_StatCan_prod
```

```python
#arrange as multi-index of year (REF_DATE), geography (GEO), flow (Variable), and product category.
#keep unit (UOM), coordinate and value as columns
#'COORDINATE':GEO.Variable.Product category
df_StatCan_prod_multiind=df_StatCan_prod.set_index(['REF_DATE','GEO','Variable','Product category'],inplace=False)
df_StatCan_prod_multiind
```

```python
#import Variabl matching with From and To, 
#compiled 2024-05-06
data_flowmatch_StatCan_prod = pd.read_csv('../Code_Test/38100150_flow_from-to_matching.csv',sep=',')
#remove unnecessary columns
data_flowmatch_StatCan_prod=pd.DataFrame(data_flowmatch_StatCan_prod, columns=['Variable','From','To'])
data_flowmatch_StatCan_prod
```

```python
#convert data_flowmatch_StatCan_prod into dictionary to match From and To in df_StatCan_prod_multiind:
#df_dict_flowmatch=data_flowmatch_StatCan_prod.to_dict(orient='records')
#df_dict_flowmatch
```

```python
#merge data frames
#reset index
regular_df_StatCan_prod = df_StatCan_prod_multiind.reset_index()
merged_df_StatCan_prod = pd.merge(regular_df_StatCan_prod, data_flowmatch_StatCan_prod, on='Variable', how='outer',validate='many_to_many')
merged_df_StatCan_prod
```

```python
#recover multi-index
df_StatCan_prod_multiind=merged_df_StatCan_prod.set_index(['REF_DATE','GEO','Variable','Product category'],inplace=False)

#save df to Excel
#df_StatCan_prod_multiind.to_excel('38100150_df.xlsx')
```
