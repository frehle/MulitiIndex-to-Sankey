```python
import pandas as pd
import numpy as np
```

```python
#import raw Statistics Canada physical flow account for plastic material, by resin, 
#from .csv, release date: 2024-03-18
data_raw_StatCan_resin = pd.read_csv('38100151.csv',sep=',')
```

```python
data_raw_StatCan_resin
```

```python
#create DataFrame from raw data
df_StatCan_resin=pd.DataFrame(data=data_raw_StatCan_resin)
df_StatCan_resin
```

```python
#remove unnecessary columns
df_StatCan_resin=pd.DataFrame(df_StatCan_resin, columns=['REF_DATE','GEO','Variable','Resin',
                                                       'UOM','COORDINATE','VALUE'])
df_StatCan_resin
```

```python
#arrange as multi-index of year (REF_DATE), geography (GEO), flow (Variable), and Resin.
#keep unit (UOM), coordinate and value as columns
#'COORDINATE':GEO.Variable.Resin
df_StatCan_resin_multiind=df_StatCan_resin.set_index(['REF_DATE','GEO','Variable','Resin'],inplace=False)
df_StatCan_resin_multiind
```

```python
#merge data frames
#reset index
regular_df_StatCan_resin = df_StatCan_resin_multiind.reset_index()
merged_df_StatCan_resin = pd.merge(regular_df_StatCan_resin, data_flowmatch_StatCan_prod, on='Variable', how='outer',
                          validate='many_to_many')
merged_df_StatCan_resin
```

```python
#recover multi-index
df_StatCan_resin_multiind=merged_df_StatCan_resin.set_index(['REF_DATE','GEO','Variable','Resin'],inplace=False)
df_StatCan_resin_multiind
#save df to Excel
df_StatCan_resin_multiind.to_excel('38100151_df.xlsx')
```
